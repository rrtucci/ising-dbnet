# beta_jj=0.3 Efficiency Plots for Ising-like dynamical Bnet

We plot the efficiency of our 2-dim Ising-like dynamical bnet for various temperatures and beta_jj=0.3.

# p0 = .3

```python
beta_jj = BETA_JJ_CURIE * 20
jj = 1  # no jj dependence when h=0
beta = beta_jj / jj
num_iter = 6

p0 = .3
h=0
lam=0
param_to_x_y = dict()
net = Net(beta=beta, 
          jj=jj, 
          h=h, 
          lam=lam, 
          num_iter=num_iter, 
          p0=p0,
          do_reversing=False) #no change if do reversing
net.plot_lattice("test.txt")
param_to_x_y[beta_jj]= net.get_av_cond_info_and_entropy()
```
